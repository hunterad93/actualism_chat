```python
import json
import os

# Load the JSON data
with open('/Users/adamhunter/Documents/misc/actualism_chat/actualfreedom/output.json', 'r') as file:
    data = json.load(file)

# Dictionary to hold the longest text for each URL
filtered_data = {}

for entry in data:
    url = entry['url']
    text = entry['text']
    if url in filtered_data:
        # If the URL already exists, keep the longer text
        if len(text) > len(filtered_data[url]['text']):
            filtered_data[url] = entry
    else:
        # If the URL does not exist, add the entry
        filtered_data[url] = entry

# Create a directory to store the output files
output_dir = 'output_files'
os.makedirs(output_dir, exist_ok=True)

# Write the filtered data to files
for url, entry in filtered_data.items():
    filename = url.replace('http://', '').replace('/', '---') + '.txt'
    with open(os.path.join(output_dir, filename), 'w') as text_file:
        text_file.write(entry['text'])

    print(f"File saved: {filename}")
```
